##  Konﬂikt: Qualität vs. Quantität

Fast immer, wenn die Forderung gestellt wird, Daten zu komprimieren, um mehr Datenmengen speichern zu können, ist damit ein Qualitätsverlust verbunden. Auch beim mp3 Verfahren war es die Aufgabe Speicherplatz z.B. von Musik einzusparen. Die Herausforderung war es, einen Song nur soweit zu komprimieren, dass er immer noch eine gute Qualität aufweist und man als Hörer nicht nur Rauschen wahrnimmt. Dazu wurde ein komplexes Hörmodell entwickelt, welches die Töne eines Songs, welche der Mensch gar nicht oder sehr gering wahrnehmen kann, aus einem Song raus streicht.

# MP3 - Aufgabenblatt 1: Einen Ton mathematisch modellieren
## Einleitung

Jedes Instrument und auch die menschliche Stimme verursachteinegewisseArtvonSchwingung,auchwenndiese für normale Menschen nicht wahrnehmbar ist und nur mit Hilfe eines Computers oder eines Oszilloskops sichtbar gemacht werden kann. So ist beispielsweise der Ton einer Stimmgabel mathematisch bzw. physikalisch nichts anderes als eine harmonische Schwingung in Form einer Sinuskurve und kann mit der Sinusfunktion 

$$g(t) = A \cdot sin(2 \cdot π \cdot f \cdot  t)$$

modelliert werden. Hier seht ihr den (Teil-)Graphen einer Sinusschwingung:                                                                                                                                                                                                                                                           
![Stimmgabel.png](../MP3_Bilder/Stimmgabel.png)

Dabei steht

- $t$ für die **Zeit**.
- $A$ für die **Amplitude** und gibt die maximale **Lautstärke** des Tons an. Das heißt je höher die Amplitude, desto lauter wird der Ton wahrgenommen.
- $f$ für die **Frequenz** und gibt die **Tonhöhe** an. der Ab stand zwischen zwei gleichen Maxima wird als Periode $T = 1/f$ bezeichnet. D.h. je größer die Frequenz $f$ ist, desto näher liegen die Maxima nebeneinander und umso höher wird der Ton wahrgenommen. Die Frequenz wird in Hertz [Hz] angegeben.
